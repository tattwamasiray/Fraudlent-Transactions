## Create 2 random samples based on raw data fraud.csv
### These will be used to upload to the website by the end user

```python
# Import Dependencies
import random
import pandas as pd
```

```python
# Load existing data
fraud_df = pd.read_csv('Fraud.csv')
# Drop the second column that was the original index column
fraud_df.drop(fraud_df.columns[0], axis=1, inplace=True)

# Create two random samples of 300 rows each
num_samples = 2
sample_size = 300
random_state = 2023

samples = [fraud_df.sample(n=sample_size, random_state=random_state) for _ in range(num_samples)]
```

```python
# Check first sample 
samples[0]
```

```python
# Check second sample 
samples[1]
```

```python
# Create an independent DataFrame from the first sample
sample1_df = samples[0].copy()  

# Create an independent DataFrame from the second sample
sample2_df = samples[1].copy()  
```

```python
# Show the proportion of the fraudulant and non-fraudulant transactions in sample_1_df
sample1_df['is_fraud'].value_counts()
```

```python
# Show % proportion of the fraudulant and non-fraudulant transactions in sample_1_df
sample1_df['is_fraud'].value_counts(normalize=True)
```

```python
# Show the proportion of the fraudulant and non-fraudulant transactions in sample_2_df
sample2_df['is_fraud'].value_counts()
```

```python
# Show % proportion of the fraudulant and non-fraudulant transactions in sample_2_df
sample2_df['is_fraud'].value_counts(normalize=True)
```

```python
# Drop is_fraud column from sample files
sample1_df = sample1_df.drop("is_fraud", axis=1)
sample2_df = sample2_df.drop("is_fraud", axis=1)
```

```python
# Export and save sample dataframes into csv files to be uploaded to web application
sample1_df.to_csv('sample1.csv', index=False)
sample2_df.to_csv('sample2.csv', index=False)
```
